## Top Performing Schools (By % Overall Passing)

* Sort and display the top five performing schools by % overall passing.

```python
df["passing_math"] = (df['math_score'])>=70
df["passing_reading"] = (df.reading_score)>=70
df["overall_pass"] = df['passing_math'] & df['passing_reading']

group_by_school = df.groupby('school_name')
math_score = group_by_school['math_score'].mean()
reading_score = group_by_school['reading_score'].mean()
total_students = group_by_school['student_name'].count()
school_type = school_data.set_index(["school_name"])["type"]
total_budget = school_data.set_index(["school_name"])["budget"]
per_student_budget = total_budget/total_students


passing_math_percent = group_by_school['passing_math'].apply(lambda x:np.sum(x)/len(x)*100)
passing_reading_percent = group_by_school['passing_reading'].apply(lambda x:np.sum(x)/len(x)*100)
passing_overall_percent = group_by_school['overall_pass'].apply(lambda x:np.sum(x)/len(x)*100)


summary_by_school = pd.DataFrame({"School Type": school_type,
                                  "Total Students": total_students,
                                  "Total School Budget": total_budget.map('${:,.2f}'.format),
                                  "Per Student Budget": per_student_budget.map('${:.2f}'.format),
                                  "Average Math Score": math_score,
                                  "Average Reading Score": reading_score,
                                  "% Passing Math": passing_math_percent,
                                  "% Passing Reading": passing_reading_percent,
                                  "% Overall Passing": passing_overall_percent,
                                  
                              }).rename_axis(None)
summary_by_school
```
